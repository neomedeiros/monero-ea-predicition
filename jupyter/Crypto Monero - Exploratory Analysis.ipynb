```python
import warnings
warnings.filterwarnings('ignore')

import sklearn.model_selection
import sklearn.datasets
import sklearn.metrics
import autosklearn.regression
import pandas
from matplotlib import pyplot
```

```python
dfxmr = pandas.read_csv('/home/leandro/projects/cryptobot/exploratoryAnalysis/datasets/crypto_datasets/xmr_complete.csv')
```

```python
dfxmr.head()
```

```python
pyplot.figure(figsize=(17,8))
pyplot.plot(dfxmr['index'], dfxmr['price(USD)'])
```

```python
pyplot.figure(figsize=(17,8))
pyplot.scatter(x=dfxmr['index'], y=dfxmr['price(USD)'])
```

```python
dfxmr.shape
```

```python
x = dfxmr.loc[:,['index','txCount','exchangeVolume(USD)']]
```

```python
x.shape
```

```python
y = dfxmr['price(USD)']
```

```python
y.shape
```

```python
X_train, X_test, y_train, y_test = \
        sklearn.model_selection.train_test_split(x, y, random_state=1)
```

```python
pyplot.figure(figsize=(17,8))
pyplot.scatter(x=y_test.to_frame().index,y=y_test.to_frame()['price(USD)'])
```

```python
automl = autosklearn.regression.AutoSklearnRegressor(
        time_left_for_this_task=120,
        per_run_time_limit=30                
    )
```

```python
feature_types = (['numerical'] + ['numerical'] + ['Numerical'])
```

```python
automl.fit(X=X_train,y=y_train, dataset_name='test',feat_type=feature_types)
```

```python
print(automl.show_models())
```

```python
predictions = automl.predict(X_test)
```

```python
print("MAE score:", sklearn.metrics.mean_absolute_error(y_test, predictions))
print('R2 score',  sklearn.metrics.r2_score(y_test, predictions))
```

```python
pyplot.figure(figsize=(17,8))
pyplot.scatter(x=y_test.to_frame().index,y=y_test.to_frame()['price(USD)'], label='Test Sample')
pyplot.scatter(x=y_test.to_frame().index,y=predictions, label="Prediction Data")
pyplot.legend(loc='upper center', shadow=True, fontsize='x-large')
```
